```python
import itertools

map = {"UUU":"F", "UUC":"F", "UUA":"L", "UUG":"L",
"UCU":"S", "UCC":"s", "UCA":"S", "UCG":"S",
"UAU":"Y", "UAC":"Y", "UAA":"STOP", "UAG":"STOP",
"UGU":"C", "UGC":"C", "UGA":"STOP", "UGG":"W",
"CUU":"L", "CUC":"L", "CUA":"L", "CUG":"L",
"CCU":"P", "CCC":"P", "CCA":"P", "CCG":"P",
"CAU":"H", "CAC":"H", "CAA":"Q", "CAG":"Q",
"CGU":"R", "CGC":"R", "CGA":"R", "CGG":"R",
"AUU":"I", "AUC":"I", "AUA":"I", "AUG":"M",
"ACU":"T", "ACC":"T", "ACA":"T", "ACG":"T",
"AAU":"N", "AAC":"N", "AAA":"K", "AAG":"K",
"AGU":"S", "AGC":"S", "AGA":"R", "AGG":"R",
"GUU":"V", "GUC":"V", "GUA":"V", "GUG":"V",
"GCU":"A", "GCC":"A", "GCA":"A", "GCG":"A",
"GAU":"D", "GAC":"D", "GAA":"E", "GAG":"E",
"GGU":"G", "GGC":"G", "GGA":"G", "GGG":"G"}

#as we can find 2 items with the same value but different keys. 
#this function inverts the map and stores the now numerous values belonging
#to the same key in an array 
def invertDict(map):
    
    inv_map = {}
    
    values = set(map.values()) #get all the values in the map 
    
    for value in values:
        
        list = [] #used to store different keys with the same values 
        
        for key in map:
            
            if map[key] == value:
                
                list.append(key)
                
        if len(list) >= 1:
            
            inv_map[value] = list
            
        else:
            
            inv_map[value] = list[0]
            
    return inv_map #the different codons that refers to the same amino acid are now stored in a list

inverted_map = invertDict(map)

def revTrans(proteinSeq):
    
    all_list = [] #store different codons 
    
    for protein in proteinSeq:
        
        all_list.append(inverted_map[protein])
        
    all_possible_comb = list(itertools.product(*all_list)) #permutate
    
    print("The Number of possible combination is: ", len(all_possible_comb))
    
    str_comb = [''.join(i) for i in all_possible_comb]#convert tuples to string
    
    #print (str_comb)
    
    print("List of all possible mRNA from lowest %GC to highest: ")
    
    perc_list = []
    
    dict1= {}
    
    dict2 = {}
    
    close_50 = []
    
    for j in range(len(str_comb)):
        
        seq = str_comb[j]
        
        countCG = seq.count("G") + seq.count("C") #count the number of Gs and Cs in the sequence
        
        percGC = round((countCG / len(seq) * 100),2) #compute %GC
        
        perc_list.append(percGC)
        
        close = abs(50 - percGC)
        
        close_50.append(close)
        
        for x in range(len(perc_list)):
            
            dict1[seq] = perc_list[x]
            
            dict2[seq] = close_50[x]
    
    perc_list.sort() #Sort percentages from lowest to highest 
        
    set1 = set(perc_list) #get unique value 
        
    lst = list(set1)
    
    lst.sort()#sort list again

    for y in range(len(lst)):
            
        for key, value in dict1.items():
                
            if value == lst[y]:
                    
                print(key,": %GC =", value,"%") #print from lowest %GC to highest
                
    print('\n')
                
    print("The most probable mRNA closest to 50:")
    
    for key1, value2 in dict2.items():
            
        if value2 == min(close_50):
            
            print(key1) #print closest to 50 
            

proteinSeq = str(input("Enter a protein sequence: "))

revTrans(proteinSeq)
```
